```
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
pred_proba = pd.read_table('tmp/news_aggregator_output_predict/test_results.tsv', names=['b', 'e', 'm', 't'])
pred = pred_proba.apply(lambda x: pred_proba.columns[np.argsort(x)[-1]], axis=1)
true = pd.read_table('data/test_label.tsv', header=None)
```

```
accuracy_score(true,pred)
```
